```python
import matplotlib.pyplot as plt
import math

# target movement parameters
target0=0

# eye parameters
eye0=0
eye_v=0.5

# data to be gathered
time_data : list = [0]
target_data : list = [0]
eye_data : list = [0]

# simulation parameters
t_step=0.1
t_stop=100

# initial state
t=0
target=target0
eye=eye0

while t<t_stop:
    t=t+t_step
    target=target0+math.sin(t)
    error=target-eye
    if error<=eye_v*t_step and error>=-eye_v*t_step:
        eye=eye+error #caught up
    elif error>eye_v*t_step:
        eye=eye+eye_v*t_step #lagging because top speed exceeded
    elif error<-eye_v*t_step:
        eye=eye-eye_v*t_step #lagging because top speed exceeded in other direction

    time_data.append(t)
    target_data.append(target)
    eye_data.append(eye)
    
    print(t)

print(target_data)
print(eye_data)
print(time_data)

plt.title('target and eye trajectory')
plt.xlabel('Time [s]')
plt.ylabel('Position [deg]')
plt.plot(time_data, target_data)
plt.plot(time_data, eye_data)
plt.legend(('target', 'eye'))
plt.show()
```
